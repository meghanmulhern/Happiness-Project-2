```python
country_regions = pd.DataFrame(country_region_first.groupby('Country').first()[["Region"]].reset_index())
country_data_df_renamed = pd.merge(country_region_first.drop("Region", axis = 1), country_regions, how = "left")
#first command, grab first region, then do a merge and take first region and overwrite what is currentl there
#will create df of just country and region
country_data_df_renamed
```

```python
country_first = pd.DataFrame(country_data_df_renamed[['Country','Region']])
country_second = country_first.dropna()
country = country_second.rename(columns={'Country':'country', 'Region':'region'})
country.head()
country = country.drop_duplicates()

# null_columns=happiness_data.columns[happiness_data.isnull().any()]
# happiness_data[null_columns].isnull().sum()
# print(happiness_data[happiness_data.isnull().any(axis=1)][null_columns].head())

```

```python
country_index = pd.DataFrame(country.reset_index())
country_index.head()
country_id_df = country_index.rename(columns = { "index": "country_id"})
country_id_df.head()
```

```python
happiness_data_first = pd.DataFrame(full_data_df_renamed[['country','year', 'happiness_score','gdp', 'life_expectancy','freedom_score','trust_score','generosity_score']])
happiness_data_second = happiness_data_first.dropna()
# happiness_data_second['country_id'] = ""
# happiness_data_second["country_id"] = pd.to_numeric(happiness_data_second["country_id"])

# happiness_data_third = happiness_data_second[['country','year', 'happiness_score','gdp', 'life_expectancy','freedom_score','trust_score','generosity_score']]
happiness_data_second.head()
```

```python
happiness_data_third = pd.merge(country_id_df, happiness_data_second, how="outer", on="country")
happiness_data_four = happiness_data_third.dropna()
happiness_data = happiness_data_four[['year', 'country_id','happiness_score','gdp', 'life_expectancy','freedom_score','trust_score','generosity_score']]
happiness_data.head()
```
